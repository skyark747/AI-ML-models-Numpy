```python
import heapq
    
def create_path(path,parent):
    new_path=""
    for i in range(1,len(path)):                        # checks if the i-1 value in path taken is parent of ith, value if it is not then that
        p=path[i-1]                                     # means i-1 value was only pushed to list due property of minheap , this i-1 value is not part
        if p==parent[path[i]][0]:                       # of the path taken so drop it.
            new_path=new_path+p+" -> "                  # e.g (A,C,B,F) B is not F's parent but only pushed to list because B was found top of queue
            new_path=new_path+path[i]+" -> "            # drop B as B is neither C's nor F's neighbor
    new_path+="end"
    return new_path



def update_best_parent(parent,neighbor,dis,node):
    try:
        value=parent[neighbor]                   #checks if a key already exists , if it does then that means node having multiple parent                                                
        if value[1]>dis:                         #update the previous parent if new parent gives better path distance
            value[1]=dis                         
    except KeyError:
        parent[neighbor]=[node,dis]



def astar(G,H,start,end):
    Visited=[] #creating a list to store the nodes which have been visited to avoid cycles
    q=[]
    path=[]
    parent={} # record parent of each node

    # features of nodes pushed on minheap  => ( f(n) , path cost , edge weight , ( h(n) , name ) )
    #  f(n) selected as basis for priority

    heapq.heappush(q,(0,0,0,(H[start],start))) # minheap for priority queue
    Visited.append(start) # mark the node as visited

    while q:
        distance,p_pathcost,_,key=heapq.heappop(q) # get the element with the shortest f(n)=g(n)+h(n)
        _,node=key
        path.append(node)  #append the nodes to path taken to find end goal
        if node == end:
             new_path=create_path(path,parent) #create path taken from start to end
             return distance,p_pathcost,new_path
        
        for weight,p_cost,dis,neighbor in G[node]: # exploring neighbors of node poped from queue
            if neighbor not in Visited:
                Visited.append(neighbor) # mark the neighbor as visited

                p_cost=p_pathcost+weight # updating the path cost for each node , previous pathcost + new edge weight

                dis=H[node]+p_cost # implementing f(n)=g(n)+h(n)

                heapq.heappush(q,(dis,p_cost,weight,(H[neighbor],neighbor)))  #push elements on the basis of f(n)

                update_best_parent(parent=parent,neighbor=neighbor,dis=dis,node=node) # update the previous parent if new better parent found
                
                
    return -1


G={
    "A":[(9,0,0,"B"),(4,0,0,"C"),(7,0,0,"D")], 
    "B":[(9,0,0,"A"),(11,0,0,"E")],
    "C":[(4,0,0,"A"),(12,0,0,"F"),(17,0,0,"E")],
    "D":[(7,0,0,"A"),(14,0,0,"F")],
    "E":[(11,0,0,"B"),(17,0,0,"C"),(5,0,0,"Z")],
    "F":[(12,0,0,"C"),(14,0,0,"D"),(9,0,0,"Z")],
    "Z":[(5,0,0,"E"),(9,0,0,"F")]
}
H={
    "A":21,
    "B":14,
    "C":18,
    "D":18,
    "E":5,
    "F":8,
    "Z":5
}

distance,pathcost,path=astar(G,H,"A","Z")

print(f"Shortest f(n) distance : {distance} , shortest path cost  : {pathcost}")
print(f"path taken = {path}")
```
