```python
import random

def print_queens(Arr,D):
    for i in range(D):
        for j in range(D):
            if j==Arr[i]:
                print('Q',end=' ')
            else:
                print('.',end=' ')
        print()

def check_vertical(S,q_n):
    for i in range(len(S)):
        if i!=q_n and S[i]==S[q_n]:
            return False
    return True

def check_diagonal(S,r):
    for i in range(len(S)):
        if i!=r and abs(i-r)==abs(S[i]-S[r]):
            return False
    return True

def population_initialization(D):
    P=[]
    for i in range(1000):
        r=[]
        for i in range(D):
            r.append(random.randint(0,D-1))
        P.append(r)
    return P


def fitness_check(P,D):
    F=[]
    for j in range(len(P)):
        score=0
        for i in range(D):
            if check_vertical(P[j],i) and check_diagonal(P[j],i):
                score+=1
        F.append(score)
    return F,P

def selection(P,F):
    S=random.choices(P,F,k=2)
    return S[0],S[1]

def cross_over(p1,p2):
    num = random.randint(0, len(p1) - 1)
    return p1[:num] + p2[num:], p2[:num] + p1[num:] 

def mutation(C1,C2,D):
    num1=random.randint(0,len(C1)-1)
    C1[num1]=random.randint(0,D-1)
    num2=random.randint(0,len(C1)-1)
    C2[num2]=random.randint(0,D-1)
    return C1,C2


Dimention=int(input("enter dimensions of chess board:"))
P=population_initialization(Dimention)
fitness=0
flag=False
optimal_value=0
max_generations=2000
gen=0
while not flag:
    gen+=1

    if gen>=max_generations:
        print("no optimal solution found")
        break
    

    F,S=fitness_check(P,Dimention)

    for i in range(len(F)):
        if F[i]==Dimention:
            flag=True
            optimal_value=S[i]
            break

    p1,p2=selection(S,F)
    
    child1,child2=cross_over(p1,p2)
    child1,child2=mutation(child1,child2,Dimention)
    P.append(child1)
    P.append(child2)  

print(f'generations taken : {gen}\n')
if gen<2000:
    print_queens(optimal_value,Dimention)
```
