```python
import random

class Node:
    def __init__(self,value):
        self.value=value
        self.left=None
        self.right=None


def createTree(depth,n):
    if depth==n:
        R=random.randint(1,500)
        return Node(R)
    new_node=Node(-1)
    new_node.left=createTree(depth+1,n)
    new_node.right=createTree(depth+1,n)
    return new_node
        
def dfs(Node):
    if Node is None:
        return
    print(Node.value)
    Node.left=dfs(Node.left)
    Node.right=dfs(Node.right)

def min_max(Node,depth,n):
    if Node is None:
        return 0
    if depth==n:
        return Node.value
    
    left=min_max(Node.left,depth+1,n)
    right=min_max(Node.right,depth+1,n)

    if (depth) % 2 == 0:
        return max(left,right)
    else:
        return min(left,right)


def minimax_pruning(Node, depth, n, alpha, beta):
    if Node is None:
        return 0  
    
    if depth == n:
        return Node.value  
    
    if depth % 2 == 0:
        left = minimax_pruning(Node.left, depth+1, n, alpha, beta)
        alpha = max(alpha, left)
        if alpha >= beta:
            return alpha 

        right = minimax_pruning(Node.right, depth+1, n, alpha, beta)
        alpha = max(alpha, right)
        if alpha >= beta:
            return alpha
        
        return alpha
        
    else:  
        left = minimax_pruning(Node.left, depth+1, n, alpha, beta)
        beta = min(beta, left)
        if alpha >= beta:
            return beta 

        right = minimax_pruning(Node.right, depth+1, n, alpha, beta)
        beta = min(beta, right)
        if alpha >= beta:
            return beta
        
        return beta

depth=int(input("enter depth of tree: "))
Tree=createTree(0,depth)
print(f'Root value before min-max: {Tree.value}')

```

```python
value=min_max(Tree,0,depth)
print("after min-max:")
print(value)
```
